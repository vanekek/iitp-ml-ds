```python
df1 = pd.read_csv('/content/iitp-ml-ds/2023/seminar-2-lin-reg/kc_house_data_train.csv', index_col=0)
```

```python
df1.drop(columns=['id','zipcode','lat','long','date'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df1['price'].values
df1.drop(columns=['price'], inplace=True)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(df1, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr_1 = LinearRegression()
lr_1.fit(X_train,y_train)
y_pred_train = lr_1.predict(X_train)
y_pred_valid = lr_1.predict(X_valid)
```

```python
LR = []
for item in sorted(list(zip(X_train.columns, list(lr_1.coef_))), reverse=True, key = lambda x: x[1]) :
    LR.append(item[0])
    print('Feature: {} -- weight: {}'.format(item[0], item[1]))
print('Intercept: ', lr_1.intercept_)
print('MSE: ', mean_squared_error(y_train, y_pred_train))
print('MSE on valid: ', mean_squared_error(y_valid, y_pred_valid))
```

## 2. Стандартизация признаков

Поскольку в линейной модели мы ищем целевую характеристику ввиде линейной комбинации признаков, то мы бы хотели чтобы величины признаков измерялись в сопоставимых масштабах. Сравните различные подходы к стандартизации признаков:

1. Стандартизация или z-scoring: (x - mean(x)) / std(x). Среднее каждого признака становится равно 0, стандартное отклонение становится равно 1.
2. Нормализация к [0, 1]: (x - min(x)) / (max(x) - min(x))
3. Логарифмирование: log(x + 1). Потенциальным недостатком первых двух подходов является необходимость подсчета статистик на тренировочной выборке и применение их на тестовой.

Обратите внимание что подсчет всех статистик mean, std, min, max должен производиться на тренировочной части данных и применяться к тестовой (потому что на практике у вас нет никакой статистики на этапе предсказание, вы делаете предсказание для конкретного 1ого дома).

Воспользуйтесь методами `sklearn.preprocessing.StandardScaler` и `sklearn.preprocessing.MinMaxScaler`. Обучайте модель линейной регрессии на тренировочной части данных, делайте предсказание и оценивайте качество на валидационной.

4. Какой из подходов дает наилучшее качество (включая отсутствие какой бы то нибыло стандартизации данных)?
5. Как меняются веса признаков в линейной модели при применении различных видов стандартизации?

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

##Z-Scoring##

```python
scaler_1 = StandardScaler()
scaler_1.fit(X_train)
X_train_scaled = scaler_1.transform(X_train)
X_valid_scaled = scaler_1.transform(X_valid)
lr_1.fit(X_train_scaled, y_train)
y_pred_valid_scaled = lr_1.predict(X_valid_scaled)
LR_standartized = []
for item in sorted(list(zip(X_train.columns, list(lr_1.coef_))), reverse=True, key = lambda x: x[1]) :
  LR_standartized.append(item[0])
  print('Feature: {} -- weight: {}'.format(item[0], item[1]))
print('Intercept: ', lr_1.intercept_)
print('MSE on scaled valid: ', mean_squared_error(y_valid, y_pred_valid_scaled))
```

```python
X_train_scaled
```

#Min-Max#

```python
scaler_2 = MinMaxScaler()
scaler_2.fit(X_train)
X_train_scaled = scaler_2.transform(X_train)
X_valid_scaled = scaler_2.transform(X_valid)
lr_1.fit(X_train_scaled, y_train)
y_pred_valid_scaled = lr_1.predict(X_valid_scaled)
LR_normalized = []
for item in sorted(list(zip(X_train.columns, list(lr_1.coef_))), reverse=True, key = lambda x: x[1]) :
  LR_normalized.append(item[0])
  print('Feature: {} -- weight: {}'.format(item[0], item[1]))
print('Intercept: ', lr_1.intercept_)
print('MSE on scaled valid: ', mean_squared_error(y_valid, y_pred_valid_scaled))
```

#Log#

```python
X_train_log = np.log(X_train + 1)
X_valid_log = np.log(X_valid + 1)
lr_1.fit(X_train_log, y_train)
y_pred_valid_scaled = lr_1.predict(X_valid_log)
LR_log = []
for item in sorted(list(zip(X_train.columns, list(lr_1.coef_))), reverse=True, key = lambda x: x[1]) :
  LR_log.append(item[0])
  print('Feature: {} -- weight: {}'.format(item[0], item[1]))
print('Intercept: ', lr_1.intercept_)
print('MSE on scaled valid: ', mean_squared_error(y_valid, y_pred_valid_scaled))
```

```python
scaler_1 = StandardScaler()
scaler_1.fit(X_train)
X_train_scaled = scaler_1.transform(X_train)
X_valid_scaled = scaler_1.transform(X_valid)
selector.fit(X_train_scaled, y_train)
LR_standartized_RFE = [x[0] for x in sorted(list(zip(X_train.columns, selector.ranking_)), reverse = False, key = lambda x: x[1])]
feature_importance.loc[0:5,'LR-standartized-RFE'] = LR_standartized_RFE[0:5]
```
